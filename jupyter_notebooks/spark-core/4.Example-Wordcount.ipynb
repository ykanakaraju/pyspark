**method - 1**

```python
file_path = "E:\\Spark\\wordcount.txt"
```

```python
wordcount_rdd = sc.textFile(file_path, 4) \
                .flatMap(lambda x: x.split()) \
                .groupBy(lambda x: x) \
                .mapValues(len) \
                .sortBy(lambda x: x[1], False, 1)
```

```python
wordcount_rdd.glom().collect()
```

```python
wordcount_rdd.getNumPartitions()
```

```python
wordcount_rdd.saveAsTextFile("E:\\PySpark\\output\\wc")
```

**method - 2**

```python
wordcount_rdd = sc.textFile(file_path, 4) \
                .flatMap(lambda x: x.split()) \
                .map(lambda x: (x, 1)) \
                .groupByKey() \
                .mapValues(sum) \
                .sortByKey(True, 1)
    
wordcount_rdd.collect()
```

**method - 3**

```python
wordcount_rdd = sc.textFile(file_path, 4) \
                .flatMap(lambda x: x.split()) \
                .map(lambda x: (x, 1)) \
                .reduceByKey(lambda x, y: x + y) \
                .sortByKey(True, 1)
        
wordcount_rdd.collect()
```
